## Pairplot
метод визуализации данных из библиотеки Seaborn, который создает матрицу графиков для сравнения всех числовых признаков в наборе данных "каждый с каждым". Он строит диаграммы рассеяния (
) для пар переменных и гистограммы (или графики плотности) для отдельных переменных по диагонали. 
Назначение: Быстрый исследовательский анализ данных (EDA), выявление корреляций, кластеров и выбросов.

```python

sns.pairplot(df, hue='species', diag_kind='kde',
             palette='husl', height=2.5,
             plot_kws={'alpha': 0.7})
plt.suptitle('Pairplot: все пары признаков', y=1.02)
plt.show()
```

## df.corr()
метод вычисления матрицы корреляции между столбцами в DataFrame.
Показывает, насколько сильно две переменные связаны друг с другом и в каком направлении меняются.
По умолчанию метод использует коэффициент корреляции Пирсона, который оценивает линейную зависимость. Результатом работы является новая таблица (матрица), где: 

1.0  идеальная прямая связь (одна переменная растет, другая тоже).

-1.0 идеальная обратная связь (одна растет, другая падает).

0    отсутствие линейной связи.

```python
corr = df[iris.feature_names].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, square=True, vmin=-1, vmax=1)
plt.title('Корреляционная матрица')
plt.show()
# TODO: Какие два признака коррелируют сильнее всего? Запишите ниже:
# Ответ: _____ и _____ (r = _____)
```

## KNN классификация

```python
from sklearn.model_selection import train_test_split

X = df[iris.feature_names].values
y = df['target'].values

# TODO: заполните параметры — 30% на тест, стратификация, random_state=42
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,       # доля тестовой выборки
    stratify=y,        # сохранить пропорции классов
    random_state=42
)
print(f"Train: {X_train.shape[0]}, Test: {X_test.shape[0]}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Шаг 1: kNN без масштабирования
knn_raw = KNeighborsClassifier(n_neighbors=5)
knn_raw.fit(X_train, y_train)
y_pred_raw = knn_raw.predict(X_test)
acc_raw = accuracy_score(y_test, y_pred_raw)
print(f"Accuracy БЕЗ масштабирования: {acc_raw:.4f}")
```

```python
from sklearn.preprocessing import StandardScaler

# Шаг 2: масштабирование — fit на TRAIN, transform на TEST!
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)   # учимся + трансформируем
X_test_sc = scaler.transform(X_test)         # ТОЛЬКО трансформируем!

knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_sc, y_train)
y_pred = knn_scaled.predict(X_test_sc)
acc_scaled = accuracy_score(y_test, y_pred)
print(f"Accuracy С масштабированием:  {acc_scaled:.4f}")
print(f"Разница: {(acc_scaled - acc_raw)*100:+.1f} п.п.")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Абсолютные числа
cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm, display_labels=iris.target_names).plot(
    cmap='Blues', ax=axes[0], values_format='d')
axes[0].set_title('Абсолютные значения')

# Нормализованная
cm_norm = confusion_matrix(y_test, y_pred, normalize='true')
ConfusionMatrixDisplay(cm_norm, display_labels=iris.target_names).plot(
    cmap='Blues', ax=axes[1], values_format='.2f')
axes[1].set_title('Доли (нормализованная)')
plt.tight_layout()
plt.show()
```

```python
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

```python
k_range = range(1, 21)
test_acc = []
for k in k_range:
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train_sc, y_train)
    test_acc.append(knn_temp.score(X_test_sc, y_test))

plt.figure(figsize=(10, 5))
plt.plot(k_range, test_acc, 'o-', linewidth=2)
plt.xlabel('k (число соседей)')
plt.ylabel('Accuracy')
plt.title('Как k влияет на точность?')
plt.grid(True, alpha=0.3)
plt.show()

best_k = list(k_range)[np.argmax(test_acc)]
print(f"Лучшее k = {best_k}, accuracy = {max(test_acc):.4f}")
```
